```python
df3=pd.merge(df,df2,on='Country Code',how='left')
df3.head(5)
```

```python
df3.Country.value_counts()
```

```python
country_names=df3.Country.value_counts().index
country_names
```

```python
country_values=df3.Country.value_counts().values
country_values
```

```python
#top 3 countries that uses zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

```python
Observation
Zomato has maximum transactions are in India, followed by USA and then UK
```

```python
df3.columns
```

```python
df3.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings=df3.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
ratings
```

```python
#Observation
1-When rating is between 4.5-4.9----> Excellent
2-When rating is between 4.0-4.4----> Very Good
3-When rating is between 3.5-3.9----> Good
4When rating is between 2.5-3.4----> Average
5When rating is between 1.8-2.4----> Poor
6- Zero rating given by 2148 people
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Aggregate rating', y='Rating Count',hue="Rating color", data=ratings,palette=["Blue","Red","Orange","Yellow","green","green"])
plt.show()
```

```python
#Observation
1-Not rated count is very high
2-Max rating between 2.5-3.4
```

```python
#count plot
sns.countplot(x="Rating color",data=ratings,palette=["Blue","Red","Orange","Yellow","green","green"])
plt.show()
```

```python
# find the countries name that has given 0 rating
df3[df3['Aggregate rating'] == 0].groupby('Country').size().reset_index()
```

```python
## Observation
1- Maximum no of 0 ratings are from Indian customers.
```

```python
# Find out which currency is used by which country
df3.groupby(['Country','Currency']).size().reset_index()
```

```python
# Which Countries have online delivery Option?
df3.groupby(['Country','Has Online delivery']).size().reset_index()
```

```python
#Observation
1-Online deliveries are available in India and UAE
```
